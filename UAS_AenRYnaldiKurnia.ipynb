```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/SupplyKantor.csv')

data_tgl = data['TanggalPesan']
data_kota = data['Kota']
data_sales = data['SalesRep']
data_jenis = data['Jenis']
data_jum = data['Jumlah']
data_harga = data['Harga Satuan']

kota_unik = data_kota.unique()

colors = ['red', 'blue', 'green', 'orange']
color_mapping = {kota: colors[i % len(colors)] for i, kota in enumerate(kota_unik)}

for kota in kota_unik:
    kota_data = data[data['Kota'] == kota]
    plt.scatter(kota_data['TanggalPesan'], kota_data['Harga Satuan'], s=50, c=color_mapping[kota], alpha=0.7, label=kota)

plt.title('Scatter Plot - Harga Satuan terhadap Tanggal Pesan')
plt.xlabel('Tanggal Pesan')
plt.ylabel('Harga Satuan')

plt.legend()

plt.xticks(rotation=90)

plt.show()
```
